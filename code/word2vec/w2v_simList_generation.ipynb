```python
with open(r'C:\Users\Mrulay\OneDrive - University of Windsor\uWindsor\COMP 8730 - NLP\Assignments\3. Lexical vs Vector Semantics\data\simSet.pkl', 'rb') as f:
    SimLexWords = pickle.load(f)
```

```python
model_list = []
for (dirpath, dirnames, filenames) in os.walk(r'C:\Users\Mrulay\OneDrive - University of Windsor\uWindsor\COMP 8730 - NLP\Assignments\3. Lexical vs Vector Semantics\models\Word2Vec\Reuters Corpus'):
    model_list.extend(filenames)
    break
```

```python
with open(r'C:\Users\Mrulay\OneDrive - University of Windsor\uWindsor\COMP 8730 - NLP\Assignments\3. Lexical vs Vector Semantics\data\simLex_similarities.json', 'r') as f:
    gold_set = json.load(f)
```

```python
sims = []
for k,v in gold_set.items():
    sims.append(v)
```

```python
for f in model_list:
    vec1 = gensim.models.Word2Vec.load(r"C:\Users\Mrulay\OneDrive - University of Windsor\uWindsor\COMP 8730 - NLP\Assignments\3. Lexical vs Vector Semantics\models\Word2Vec\Reuters Corpus\{}".format(f))
    simDict = {}
    for i, w in enumerate(SimLexWords):
        try: 
            x = vec1.wv.most_similar(w, 10)
        except:
            x = []
        simWords = [m[0] for m in x]
        simVals = [m[1] for m in x]
        tempdict = {}
        for word in sims[i]:
            if word in simWords:
                id = simWords.index(word)
                if id in range(0,3):
                    id2rank = 5
                elif id in range(3,7):
                    id2rank = 4
                elif id in range(7,10):
                    id2rank = 3
                tempdict[word] = id2rank
            else:
                tempdict[word] = 0
        simDict[w] = tempdict
            
    with open('reut-{}.json'.format(f), 'w') as fp:
        json.dump(simDict, fp, indent=4)
```
